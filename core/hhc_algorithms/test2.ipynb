```python
from repository.metaheuristic.simulated_annealing import simulated_annealing, x_sa, linear_multiplicative_cooling
import numpy as np
import random
```

```python
class LinearProgramming_X(x_sa):
    def __init__(self, x):
        super().__init__()
        self.x = x

    def cost_function(self):
        # Función objetivo a maximizar
        return  self.x**2 + 2 

    def generate_neighbor(self):
        # Implementa la generación de un vecino
        delta_x = random.uniform(-2, 2)
        
        return LinearProgramming_X(x= (self.x + delta_x))
    
    def check_restrictions(self,) -> bool:
        print('n')
        return True
    
    def __str__(self) -> str:
        return str(self.x)
    
```

```python
x = random.uniform(-100, 100)
x0 = LinearProgramming_X(x= x)
```

```python
sa = simulated_annealing(
    x0= x0,
    cooling_operator= linear_multiplicative_cooling(t_max= 100, alpha= 0.8),
    step_max= 10000,
    t_min= 0,
    t_max= 100
)
```
